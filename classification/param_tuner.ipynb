```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping
import keras_tuner as kt

def read_image_paths_and_labels(path):
    image_files = [f for f in os.listdir(path) if f.endswith('.jpg')]
    labels_file = os.path.join(path, "labels.csv")
    if not os.path.exists(labels_file):
        raise FileNotFoundError(f"Labels file not found: {labels_file}")
    labels_df = pd.read_csv(labels_file)

    label_dict = dict(zip(labels_df['filename'], labels_df['label']))

    id_groups = {}
    processed_filenames = set() # To track which filenames have labels

    for img_file in image_files:
        try:
            id_num = '_'.join(img_file.split('_')[:-1]) # Extracts "ID_AugmentationNumber"
            img_part = img_file.split('_')[-1] # Extracts "IMGX.jpg"
            img_idx_str = img_part.split('.')[0].replace('IMG', '') # Extracts "X"
            img_idx = int(img_idx_str)

        except (IndexError, ValueError):
            print(f"Warning: Skipping file with unexpected format: {img_file}")
            continue

        if id_num not in id_groups:
            id_groups[id_num] = {}

        img_path = os.path.join(path, img_file)
        id_groups[id_num][f"IMG{img_idx}"] = img_path
        processed_filenames.add(img_file) # Add to set of processed files

    data = []
    required_imgs = [f"IMG{i}" for i in range(1, 6)]

    for id_num, images_dict in id_groups.items():
        # Ensure all 5 images are present for this ID
        if all(key in images_dict for key in required_imgs):
            # Try to get label using the filename of the first image
            label_key = f"{id_num}_IMG1.jpg"
            label = label_dict.get(label_key)

            if label is not None:
                 # Ensure images are sorted correctly by index (IMG1, IMG2, ..., IMG5)
                sorted_image_paths = [images_dict[f"IMG{i}"] for i in range(1, 6)]
                row = {'ID': id_num, 'Label': label}
                row.update({f"IMG{i}": path for i, path in enumerate(sorted_image_paths, 1)})
                data.append(row)
            else:
                print(f"Warning: Label not found for ID {id_num} using key {label_key}")

    df = pd.DataFrame(data)

    for col in ['ID', 'Label'] + [f"IMG{i}" for i in range(1, 6)]:
        if col not in df.columns:
            df[col] = pd.NA 

    return df

train_path = "/home/furix/Desktop/DiscAI/data/Full Data/train"
test_path = "/home/furix/Desktop/DiscAI/data/Full Data/test"
val_path = "/home/furix/Desktop/DiscAI/data/Full Data/val"

print("Loading dataframes...")
try:
    train_df = read_image_paths_and_labels(train_path)
    test_df = read_image_paths_and_labels(test_path)
    val_df = read_image_paths_and_labels(val_path)

    print(f"Train DF length: {len(train_df)}")
    print(f"Test DF length: {len(test_df)}")
    print(f"Val DF length: {len(val_df)}")

    # Check if dataframes are empty, which might indicate path or file format issues
    if train_df.empty or test_df.empty or val_df.empty:
         print("\nWarning: One or more dataframes are empty. Check paths and file structures.")
         print(f"Train path exists: {os.path.exists(train_path)}")
         print(f"Test path exists: {os.path.exists(test_path)}")
         print(f"Val path exists: {os.path.exists(val_path)}")


    print("\nTrain Label Distribution:")
    print(train_df['Label'].value_counts())
    print("\nValidation Label Distribution:")
    print(val_df['Label'].value_counts())
    print("\nTest Label Distribution:")
    print(test_df['Label'].value_counts())


except FileNotFoundError as e:
    print(f"Error loading data: {e}")
    print("Please ensure the paths are correct and the 'labels.csv' file exists in each directory.")

    exit()

except Exception as e:
    print(f"An unexpected error occurred during data loading: {e}")
    exit()


# %%
INPUT_SHAPE = (30, 50, 1)
IMG_SIZE = (30, 50) 
NUM_INPUTS = 5
BATCH_SIZE = 32 

# %%
def load_and_process_image(path):
    try:
        image = tf.io.read_file(path)
        image = tf.image.decode_jpeg(image, channels=1)
        image = tf.image.resize(image, IMG_SIZE)
        image = tf.cast(image, tf.float32) / 255.0 
        return image
    except tf.errors.NotFoundError:
        # Handle cases where an image file path might be invalid or missing
        print(f"Warning: Could not find file {path.numpy().decode()}, returning zeros.")
        return tf.zeros(INPUT_SHAPE, dtype=tf.float32)
    except Exception as e:
        # Catch other potential decoding/processing errors
        print(f"Warning: Error processing file {path.numpy().decode()}: {e}. Returning zeros.")
        return tf.zeros(INPUT_SHAPE, dtype=tf.float32)

def process_sample(img_paths, label):

    images = tf.stack([load_and_process_image(p) for p in tf.unstack(img_paths)], axis=0)
    # Ensure the shape is correct after stacking and processing
    images.set_shape((NUM_INPUTS, INPUT_SHAPE[0], INPUT_SHAPE[1], INPUT_SHAPE[2]))

    # Unpack images for the model input structure (list/tuple of tensors)
    img1, img2, img3, img4, img5 = tf.unstack(images, num=NUM_INPUTS, axis=0)
    return (img1, img2, img3, img4, img5), tf.cast(label, tf.float32)


def build_dataset(df, batch_size=BATCH_SIZE, shuffle=True):

    if df.empty:
        print("Warning: Input DataFrame is empty. Cannot build dataset.")
        # Return an empty dataset to avoid errors downstream
        return tf.data.Dataset.from_tensor_slices((
             tf.zeros((0, NUM_INPUTS), dtype=tf.string), # Placeholder for paths
             tf.zeros((0,), dtype=tf.float32) # Placeholder for labels
             )).batch(batch_size)

    img_path_cols = [f'IMG{i}' for i in range(1, 6)]
    # Check if all required image columns exist
    if not all(col in df.columns for col in img_path_cols):
         raise ValueError(f"DataFrame is missing required image path columns: {img_path_cols}")

    # Ensure labels are present and numeric
    if 'Label' not in df.columns:
        raise ValueError("DataFrame is missing the 'Label' column.")
    if not pd.api.types.is_numeric_dtype(df['Label']):
        print("Warning: 'Label' column is not numeric. Attempting conversion.")
        df['Label'] = pd.to_numeric(df['Label'], errors='coerce')
        # Drop rows where conversion failed
        df.dropna(subset=['Label'], inplace=True)
        if df.empty:
            print("Warning: DataFrame became empty after handling non-numeric labels.")
            return tf.data.Dataset.from_tensor_slices((
                tf.zeros((0, NUM_INPUTS), dtype=tf.string),
                tf.zeros((0,), dtype=tf.float32)
                )).batch(batch_size)


    img_paths = df[img_path_cols].values
    labels = df['Label'].values.astype('float32')

    dataset = tf.data.Dataset.from_tensor_slices((img_paths, labels))

    # Use a wrapper for map to handle potential errors during file loading/processing
    def safe_process_sample(paths, lbl):
        try:
            return process_sample(paths, lbl)
        except Exception as e:
            print(f"Error in process_sample: {e}. Skipping sample.")

            return process_sample(paths, lbl) 


    dataset = dataset.map(safe_process_sample, num_parallel_calls=tf.data.AUTOTUNE)

    if shuffle:
        dataset = dataset.shuffle(buffer_size=len(df))

    dataset = dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return dataset

print("\nBuilding datasets...")
try:
    train_dataset = build_dataset(train_df, batch_size=BATCH_SIZE, shuffle=True)
    val_dataset = build_dataset(val_df, batch_size=BATCH_SIZE, shuffle=False)
    test_dataset = build_dataset(test_df, batch_size=BATCH_SIZE, shuffle=False)
    print("Datasets built successfully.")

except ValueError as e:
    print(f"Error building datasets: {e}")
    exit()

except Exception as e:
    print(f"An unexpected error occurred during dataset building: {e}")
    exit()


# %%

print("\nSetting up Keras Tuner...")

def build_model_for_tuner(hp):
    """
    Builds and compiles the Keras model with hyperparameters for Keras Tuner.
    Args:
        hp: KerasTuner HyperParameters object.
    Returns:
        A compiled Keras model.
    """
    inputs = [tf.keras.Input(INPUT_SHAPE, name=f'input_{i+1}') for i in range(NUM_INPUTS)]

    # --- Hyperparameters for Convolutional Blocks ---
    conv1_filters = hp.Int('conv1_filters', min_value=8, max_value=32, step=8)
    conv2_filters = hp.Int('conv2_filters', min_value=8, max_value=32, step=8)
    conv3_filters = hp.Int('conv3_filters', min_value=16, max_value=64, step=16)
    conv_activation = hp.Choice('conv_activation', ['relu', 'elu'])
    spatial_dropout_1_rate = hp.Float('spatial_dropout_1_rate', min_value=0.1, max_value=0.4, step=0.05) # Finer steps
    spatial_dropout_2_rate = hp.Float('spatial_dropout_2_rate', min_value=0.2, max_value=0.5, step=0.05) # Finer steps
    # kernel_size_1 = hp.Choice('kernel_size_1', [(3,3), (5,5)]) # Example if tuning kernel size
    # dilation_rate_conv3 = hp.Int('dilation_rate_conv3', min_value=1, max_value=3) # Example if tuning dilation

    processed_inputs = []
    for i in range(NUM_INPUTS):
        x = inputs[i]
        # Block 1
        x = tf.keras.layers.Conv2D(
            filters=conv1_filters, kernel_size=(3,3), strides=(1,1), padding='same',
            activation=conv_activation)(x)
        x = tf.keras.layers.BatchNormalization()(x)
        # Block 2
        x = tf.keras.layers.Conv2D(
            filters=conv2_filters, kernel_size=(5,5), strides=(1,1), padding='same',
            activation=conv_activation)(x)
        x = tf.keras.layers.SpatialDropout2D(spatial_dropout_1_rate)(x)
        x = tf.keras.layers.MaxPooling2D(pool_size=2)(x)
        x = tf.keras.layers.BatchNormalization()(x)
        # Block 3
        x = tf.keras.layers.Conv2D(
            filters=conv3_filters, kernel_size=(3, 3), padding='same',
            dilation_rate=2, # Fixed dilation rate as per original model, or tune with hp.Int
            activation=conv_activation)(x)
        x = tf.keras.layers.SpatialDropout2D(spatial_dropout_2_rate)(x)
        x = tf.keras.layers.MaxPooling2D(pool_size=2)(x)
        x = tf.keras.layers.BatchNormalization()(x)
        # Global Pooling
        x = tf.keras.layers.GlobalMaxPooling2D()(x)
        processed_inputs.append(x)

    # Concatenate features
    f = tf.keras.layers.Concatenate()(processed_inputs)

    # --- Hyperparameters for Dense Head ---
    dense1_units = hp.Int('dense1_units', min_value=16, max_value=64, step=16) # Smaller range?
    dense2_units = hp.Int('dense2_units', min_value=8, max_value=32, step=8)   # Smaller range?
    dense_activation = hp.Choice('dense_activation', ['relu', 'elu', 'tanh'])
    dense_dropout_1_rate = hp.Float('dense_dropout_1_rate', min_value=0.2, max_value=0.6, step=0.1)
    dense_dropout_2_rate = hp.Float('dense_dropout_2_rate', min_value=0.2, max_value=0.6, step=0.1)
    l2_rate = hp.Float('l2_rate', min_value=1e-5, max_value=1e-3, sampling='log')
    learning_rate = hp.Float("learning_rate", min_value=1e-4, max_value=1e-2, sampling="log")

    # Dense Block 1
    f = tf.keras.layers.Dense(units=dense1_units, kernel_regularizer=tf.keras.regularizers.l2(l2_rate))(f)
    f = tf.keras.layers.BatchNormalization()(f)
    f = tf.keras.layers.Activation(dense_activation)(f)
    f = tf.keras.layers.Dropout(dense_dropout_1_rate)(f)
    # Dense Block 2
    f = tf.keras.layers.Dense(units=dense2_units, kernel_regularizer=tf.keras.regularizers.l2(l2_rate))(f)
    f = tf.keras.layers.BatchNormalization()(f)
    f = tf.keras.layers.Activation(dense_activation)(f)
    f = tf.keras.layers.Dropout(dense_dropout_2_rate)(f)
    # Output Layer
    output = tf.keras.layers.Dense(1, activation='sigmoid', name='output')(f) # Binary classification

    model = tf.keras.Model(inputs=inputs, outputs=output)

    # Compile the model
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
        loss='binary_crossentropy',
        metrics=['accuracy'] # Add F1 score or others if needed and properly configured
    )
    return model

# Instantiate the Tuner
tuner = kt.BayesianOptimization(
    hypermodel=build_model_for_tuner,
    objective='val_accuracy',  # Optimize for validation accuracy
    max_trials=15,             # Number of hyperparameter combinations to try (adjust as needed)
    executions_per_trial=2,    # Train each model once per trial (can increase for stability)
    directory='keras_tuner_dir_discai', # Directory to store results
    project_name='discai_multi_input_cnn_tuning',
    overwrite=True # Set to False to resume previous tuning
)

tuner.search_space_summary()

search_early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5, # Stop a trial early if val_loss doesn't improve for 5 epochs
    verbose=1,
    restore_best_weights=True # Restore weights from the epoch with the best val_loss
)

# --- Run the Hyperparameter Search ---
print("\nStarting hyperparameter search...")

if train_df.empty or val_df.empty:
    print("Cannot start search: Training or Validation DataFrame is empty.")
else:
    tuner.search(
        train_dataset,
        epochs=15, # Max epochs *per trial* in the search (use EarlyStopping)
        validation_data=val_dataset,
        callbacks=[search_early_stopping]
    )
    print("\nHyperparameter search finished.")

    # --- Get the Best Hyperparameters ---
    try:
        best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
        print("\n--- Best Hyperparameters Found ---")
        for param, value in best_hps.values.items():
            print(f"{param}: {value}")

        # --- Build the Best Model ---
        print("\nBuilding the best model with optimal hyperparameters...")
        best_model = tuner.hypermodel.build(best_hps)
        best_model.summary()

        # --- Train the Best Model on Full Data for More Epochs ---
        print("\nTraining the best model on the full training data...")

        # Define Early Stopping and optional LR scheduling for the final training
        final_early_stopping = EarlyStopping(
            monitor='val_loss',
            patience=10, # Allow more patience for final training
            verbose=1,
            restore_best_weights=True
        )

        history = best_model.fit(
            train_dataset,
            epochs=50, # Train for potentially more epochs now
            validation_data=val_dataset,
            callbacks=[final_early_stopping] # Add reduce_lr here if using
        )
        print("Best model training finished.")


        # %%
        # --- Evaluate the Best Model ---
        print("\nEvaluating the best model on the test dataset...")
        # Ensure test dataset is not empty
        if test_df.empty:
             print("Cannot evaluate: Test DataFrame is empty.")
             test_loss, test_accuracy = None, None
        else:
            test_loss, test_accuracy = best_model.evaluate(test_dataset)
            print(f'Test Loss: {test_loss}')
            print(f'Test Accuracy: {test_accuracy}')
            # Add F1 evaluation if metric was included and calculated


        # %%
        # --- Save the Best Model ---
        print("\nSaving the best model and weights...")
        best_model.save("discai_best_model_tuned.keras")
        best_model.save_weights("discai_best_weights_tuned.weights.h5")
        print("Model saved successfully.")

        # %%
        # --- Plot Training History of the Best Model ---
        print("\nPlotting training history...")
        plt.figure(figsize=(12, 5))

        plt.subplot(1, 2, 1)
        plt.plot(history.history['loss'], label='Train Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.title('Best Model: Training vs Validation Loss')
        plt.legend()
        plt.grid(True)

        plt.subplot(1, 2, 2)
        plt.plot(history.history['accuracy'], label='Train Accuracy')
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.title('Best Model: Training vs Validation Accuracy')
        plt.legend()
        plt.grid(True)

        plt.tight_layout()
        plt.savefig("discai_training_history_tuned.png") # Save the plot
        plt.show()

    except (ValueError, IndexError) as e:
        print(f"\nError retrieving best hyperparameters or building model: {e}")
        print("This might happen if the search did not complete successfully or found no valid trials.")


if train_df.empty or val_df.empty:
     print("\nSkipped tuner search, model training, evaluation, and saving due to empty datasets.")

print("\nScript finished.")
```
